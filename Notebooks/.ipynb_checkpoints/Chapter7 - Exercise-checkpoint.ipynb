# 6
## (a)

```python
wage = pd.read_csv('../Data/Wage.csv')
```

```python
y = wage['wage']
X = wage['age']
```

```python
regr = LinearRegression()
scores = []

for i in range(1,11):
    poly = PolynomialFeatures(i)
    X_poly = poly.fit_transform(X.values.reshape(-1,1))
    kf_10 = KFold(n_splits=10, shuffle=True)
    score = -1*cross_val_score(regr, X_poly, y, cv=kf_10,
                           scoring='neg_mean_squared_error').mean()
    scores.append(score)
```

```python
min_score = np.min(scores)
std_score = np.std(scores)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))
ax.plot(np.arange(1,11),scores)
ax.set_title("CV-Error for polynomial regressions")
ax.set_xlabel("Degree of polynomials")

ax.axhline(y = min_score + 0.2*std_score, alpha=1, c="orange",
          linestyle='--', label="0.2 std line")
ax.legend();
```

From this plot, **degree 3** is optimal degree for the polynomial.

```python
fit_list = []

# Get fit_list which contains fitted model for order 1,2,...,10
for i in range(1,11):
    X_ = PolynomialFeatures(i).fit_transform(X.values.reshape(-1,1))
    fit_ = sm.GLS(y, X_).fit()
    
    fit_list.append(fit_)

sm.stats.anova_lm(*fit_list,typ=1)
```

From ANOVA, **degree 3** is the optimal degree for the polynomial.

```python
fig, ax = plt.subplots(1,1, figsize=(8,6))
ax.scatter(X, y, facecolor='None', edgecolor='k', alpha=0.3)
sns.regplot(X, y, order=3, truncate=True, scatter=False, ax=ax);
```

## (b)

```python
scores = []
for cut_number in range(2,11):
    df_cut, bins = pd.cut(X, cut_number, retbins=True, right=True)
    df_steps = pd.concat([X, df_cut, y], keys=['age', 'age_cuts', 'wage'], axis=1)
    df_steps_dummies = pd.get_dummies(df_steps['age_cuts'], drop_first=True)
    regr = LinearRegression()
    kf_10 = KFold(n_splits=10, shuffle=True)
    score = -1*cross_val_score(regr, df_steps_dummies, y, cv=kf_10,
                           scoring='neg_mean_squared_error').mean()
    scores.append(score)
```

```python
min_score = np.min(scores)
std_score = np.std(scores)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))
ax.plot(np.arange(2,11),scores)
ax.set_title("CV-Error for step regression")
ax.set_xlabel("number of cuts")

ax.axhline(y = min_score + 0.2*std_score, alpha=1, c="orange",
          linestyle='--', label="0.2 std line")
ax.legend();
```

According to CV, Optimal number of cuts is **8**

```python
# Cut 8 times and fit linear regression to data.
df_cut, bins = pd.cut(X, 8, retbins=True, right=True)
df_steps = pd.concat([X, df_cut, y], keys=['age', 'age_cuts', 'wage'], axis=1)
df_steps_dummies = pd.get_dummies(df_steps['age_cuts'], drop_first=True)
regr = LinearRegression()
regr.fit(df_steps_dummies, y);
```

```python
# Put the test data(age_grid) and make it a X_test matrix for prediction
age_grid = np.arange(X.min(), X.max()).reshape(-1,1)
bin_mapping = np.digitize(age_grid.ravel(), bins)
X_test = pd.get_dummies(bin_mapping, drop_first=True);
```

```python
pred = regr.predict(X_test)
```

```python
# Plot
fig, ax = plt.subplots(1,1,figsize=(8,6))

ax.scatter(X, y, facecolor='None', edgecolor='k', alpha=0.3)
ax.plot(age_grid, pred, c='b')
ax.set_xlabel('age')
ax.set_ylabel('wage')
ax.set_ylim(ymin=0);
```

```python
RSS_list = []
for i in range(1,11):
    # Plot
    fig, ax = plt.subplots(1,1, figsize=(6,4))
    ax.scatter(X, y, facecolor='None', edgecolor='k', alpha=0.3)
    sns.regplot(X, y, order=i, truncate=True, scatter=False, ax=ax, ci=None)
    ax.set_title("Polynomial Order = %i" %i)
    plt.show()
    
    # Calculate RSS
    X_poly = PolynomialFeatures(i).fit_transform(X.values.reshape(-1,1))
    X_poly = sm.add_constant(X_poly)
    regr = sm.OLS(y, X_poly).fit()
    RSS = np.power(regr.resid,2).sum()
    RSS_list.append(RSS)
```

# 11
## (a)

```python
np.random.seed(1)
X1 = np.random.randn(100).reshape(-1,1)
X2 = np.random.randn(100).reshape(-1,1) * 2
epsilon = np.random.randn(100).reshape(-1,1)
y = -2.1 + 1.3 * X1 - 0.54 * X2 + epsilon
```

## (b)~(e)

```python
betas = pd.DataFrame(columns=["beta0", "beta1", "beta2"])
regr = LinearRegression()
```

```python
beta1 = 10
```

```python
# (c)
a = y - beta1 * X1
beta2 = float(regr.fit(X2,a).coef_)

# (d)
a = y - beta2 * X2
beta1 = float(regr.fit(X1,a).coef_)

beta0 = float(regr.intercept_)

betas.loc[0] = {"beta0":beta0, "beta1":beta1, "beta2":beta2}
```

```python
for i in range(1, 1000):
    # (c)
    a = y - betas["beta1"][i-1] * X1
    beta2 = float(regr.fit(X2,a).coef_)
    # (d)
    a = y - beta2 * X2
    beta1 = float(regr.fit(X1,a).coef_)
    
    beta0 = float(regr.intercept_)
    
    betas.loc[i] = {"beta0":beta0, "beta1":beta1, "beta2":beta2}
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))
for i in range(3):
    ax.scatter(range(1000),betas.iloc[:1000,i])
```

Coefficients quickly converge.

## (f)

```python
X = pd.DataFrame({"X1":X1.ravel(), "X2":X2.ravel()})
regr.fit(X,y)
beta0 = regr.intercept_
beta1 = regr.coef_.ravel()[0]
beta2 = regr.coef_.ravel()[1]
beta_ = [beta0, beta1, beta2]
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,6))
for i in range(3):
    ax.scatter(range(1000),betas.iloc[:1000,i])
    ax.axhline(y = beta_[i], alpha=1, c="red",
              linestyle='--', label="0.2 std line")
```

# 12

```python
np.random.seed(1)
X = pd.DataFrame(np.random.randn(100,100))
y = np.zeros((100,1))
epsilon = np.random.randn(100,1)
true_beta = np.random.randn(100,1)*3
y = 5 + np.dot(X,true_beta) + epsilon
```

```python
betas = []
predict_beta = pd.DataFrame({"beta":np.random.randn(100)})
regr = LinearRegression()
```

```python
for i in range(1, 1000):
    for j in range(100):
        # Fir Linear Regression
        a = y - np.dot(X.drop(j, axis=1), predict_beta.drop(j))
        regr.fit(X[j].values.reshape(-1,1),a)
        
        predict_beta.loc[j] = float(regr.coef_)
        
    tmp_beta = deepcopy(predict_beta.values)
    betas.append(tmp_beta)
```

```python
# seperate_betas = List for beta_1, beta_2, beta_3, ... , beta_100
seperate_betas = []
for beta_ in zip(*betas):
    seperate_betas.append(beta_)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))
for i in range(100):
    ax.plot(range(len(seperate_betas[i])),seperate_betas[i])
```
