# 6.1 Subset Selection
For subset selection and stepwise regression, Below codes are mainly from [pdf from Smith College](http://www.science.smith.edu/~jcrouser/SDS293/labs/2016/lab8/Lab%208%20-%20Subset%20Selection%20in%20Python.pdf)


## Figure 6.2

```python
credit = pd.read_csv('../Data/Credit.csv', index_col = 0)
```

Make categorical predictors to dummy variables

```python
dummy = pd.get_dummies(credit[['Gender','Student','Married','Ethnicity']], 
                       drop_first = True)
```

```python
credit = credit.drop(['Ethnicity','Gender','Student','Married'], axis=1).join(dummy)
```

```python
y = credit['Balance'].astype('float64')
X = credit.drop('Balance',axis=1).astype('float64')
```

```python
def process_subset(features):
    regr = sm.OLS(y, X[list(features)]).fit()
    RSS = ((regr.predict(X[list(features)]) - y) ** 2).sum()
    return {"model": regr, "RSS": RSS}

def get_best(k):
    results = []
    for combo in itertools.combinations(X.columns, k):
        results.append(process_subset(combo))
        
    models = pd.DataFrame(results)
    
    best_model = models.loc[models['RSS'].argmin()]
    
    return best_model
```

```python
models = pd.DataFrame(columns=["RSS", "model"])

for i in range(1, 12):
    models.loc[i] = get_best(i)
```

```python
models
```

If you want a summary of each model, use summary()

```python
models.loc[3,"model"].summary()
```

If you want to get $R^{2}, AIC, BIC$, Use `apply` method

```python
models.apply(lambda row: row[1].rsquared, axis=1)
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
ax1, ax2, ax3 = axes

# First Plot for AIC
aic = models.apply(lambda row: row[1].aic, axis=1)
ax1.plot(aic, "r")
ax1.plot(aic.argmin(), aic.min(), "xb", ms=10)
ax1.set_ylabel("AIC")

# Second Plot for BIC
bic = models.apply(lambda row: row[1].bic, axis=1)
ax2.plot(bic, "r")
ax2.plot(bic.argmin(), bic.min(), "xb", ms=10)
ax2.set_ylabel("BIC")

# Third Plot for Adjusted R^2
adj_r = models.apply(lambda row: row[1].rsquared_adj, axis=1)
ax3.plot(adj_r, "r")
ax3.plot(adj_r.argmax(), adj_r.max(), "bx", ms=10)
ax3.set_ylabel("Adjusted R^2")

for i in range(3):
    axes[i].set_xlabel("# Predictors")
    axes[i].set_xticks([2,4,6,8,10])
```

## Figure 6.3

```python
def process_subset(features):
    # BIC
    regr = sm.OLS(y, X[list(features)]).fit()
    bic = np.sqrt(regr.bic)

    # Validation Set Error
    regr = LinearRegression()
    X_train, X_test, y_train, y_test = train_test_split(X[list(features)],y, test_size=0.5)
    regr.fit(X_train, y_train)
    pred = regr.predict(X_test)
    validation_error = mean_squared_error(y_test, pred)
    
    # Cross-Validation Error
    regr = LinearRegression()
    kf_10 = KFold(n_splits=10)
    cv_error = -1*cross_val_score(regr, X[list(features)], y, cv=kf_10,
                           scoring='neg_mean_squared_error').mean()
    
    return {"BIC": bic, "Validation_Error": validation_error,
            "CV_Error": cv_error}

def get_best(k):
    results = []
    for combo in itertools.combinations(X.columns, k):
        results.append(process_subset(combo))
        
    results = pd.DataFrame(results)
    
    return dict(results.apply(min))
```

```python
# This will take some time.

best_results = pd.DataFrame(columns=["BIC", "Validation_Error", "CV_Error"])
for i in range(1, 12):
    best_results.loc[i] = get_best(i)
```

```python
best_results
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
ax1, ax2, ax3 = axes

# First Plot for BIC
bic = best_results['BIC']
ax1.plot(bic, "r")
ax1.plot(bic.argmin(), bic.min(), "xb", ms=10)
ax1.set_ylabel("BIC")

# Second Plot for Validation Error
val_error = best_results['Validation_Error']
ax2.plot(val_error, "r")
ax2.plot(val_error.argmin(), val_error.min(), "bx", ms=10)
ax2.set_ylabel("Validation Set Error")

# Third Plot for CV Error
cv_error = best_results['CV_Error']
ax3.plot(cv_error, "r")
ax3.plot(cv_error.argmin(), cv_error.min(), "bx", ms=10)
ax3.set_ylabel("Cross-Validation Error")

for i in range(3):
    axes[i].set_xlabel("Number of Predictors")
    axes[i].set_xticks([2,4,6,8,10])
```

## Figure 6.4
Below codes to plot Figrue 6.4 is from [Plot Ridge coefficients as a function of the regularization](http://scikit-learn.org/stable/auto_examples/linear_model/plot_ridge_path.html#sphx-glr-auto-examples-linear-model-plot-ridge-path-py)

```python
X_ = scale(X)
```

```python
# #############################################################################
# Compute paths

n_alphas = 200
alphas = np.logspace(5, -2, n_alphas)

coefs = []
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X_, y)
    coefs.append(ridge.coef_)

# #############################################################################
# Display results

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
# ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

## Figure 6.6 Lasso

In `Scikit-learn`, The optimization objective for Lasso is:

$ \dfrac{1}{2\cdot n_{samples}} \lVert y - Xw \rVert _{2} ^{2} + \alpha \lVert w \rVert _{1}$

So be careful to interpret $\alpha$ in this notebook since the interpretation of $\alpha$ is a little bit different from $\lambda$ in the book.

```python
# #############################################################################
# Compute paths

n_alphas = 200
alphas = np.logspace(3, -2, n_alphas)

coefs = []
for a in alphas:
    lasso = Lasso(alpha=a, fit_intercept=False)
    lasso.fit(X_, y)
    coefs.append(lasso.coef_)

# #############################################################################
# Display results

ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
# ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('Standardized Coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

## Figure 6.12

```python
n_alphas = 200
alphas = np.logspace(2, -3, n_alphas)

coefs = []
cv_errors = []
kf_10 = KFold(n_splits = 10)

for a in alphas:
    # CV Errors
    ridge = Ridge(alpha=a, fit_intercept=False)
    #kf_10 = KFold(n_splits = 10)
    score = -1*cross_val_score(ridge, X_, y, cv=kf_10,
                           scoring='neg_mean_squared_error').mean()
    cv_errors.append(score)
    
    # Coefficients
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X_, y)
    coefs.append(ridge.coef_)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,6))
ax1, ax2 = axes

# Left Plot
ax1.plot(alphas, cv_errors)
ax1.plot([alphas[np.argmin(cv_errors)]]*2,[min(cv_errors)-10,max(cv_errors)+10],
         linestyle='--', c='k')
ax1.set_ylabel('Cross-Validation Error')
ax1.set_ylim([min(cv_errors)-10,max(cv_errors)+10])

# Right Plot
ax2.plot(alphas, coefs)
ax2.plot([alphas[np.argmin(cv_errors)]]*2,[np.max(ridge.coef_)+10,np.min(ridge.coef_)-10],
         linestyle='--', c='k')
ax2.set_ylabel('Standardized Coefficients')
ax2.set_ylim([np.min(ridge.coef_)-10,np.max(ridge.coef_)+10])

for i in range(len(axes)):
    axes[i].set_xscale('log')
    axes[i].set_xlabel('alpha')
    #axes[i].axis('tight')
```

## Figure 6.20
`Scikit-learn` does not support PCR directly. So we should construct Principal Components by PCA. I wrote code based on [Answer to this question](https://stats.stackexchange.com/questions/82050/principal-component-analysis-and-regression-in-python)

```python
pca = PCA()
X_reduced = pca.fit_transform(scale(X))
```

```python
np.cumsum(np.round(pca.explained_variance_ratio_, 4)*100)
```

```python
coefs = []
cv_errors = []
regr = LinearRegression()
kf_10 = KFold(n_splits=10, shuffle=True)
for i in range(1,12):
    score = -1*cross_val_score(regr, X_reduced[:,:i], y, cv=kf_10,
                           scoring='neg_mean_squared_error').mean()
    cv_errors.append(score)
    
    regr.fit(X_reduced, y)
```

```python
cv_errors = pd.DataFrame(cv_errors)
cv_errors.index = range(1,12)
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(cv_errors)
ax.set_xlabel("Number of Components")
ax.set_ylabel("Cross-Validation MSE")
```
